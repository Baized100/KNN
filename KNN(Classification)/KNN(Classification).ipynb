```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import warnings as wr
wr.filterwarnings('ignore')
```

```python
df = pd.read_excel('weight-height.xlsx')
df.head()
```

```python

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df.Gender = le.fit_transform(df[['Gender']])
df.head()
```

```python
plt.scatter(df['Height'],df['Weight'], color = 'red')
plt.xlabel('Height')
plt.ylabel('Weight')
```

```python
X = df.iloc[:,:-1] 
y = df.iloc[:,-1]
```

```python

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
L_reg = LinearRegression()
```

```python
#splite data into 70% traning 30% testing
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 36)
```

```python
L_reg.fit(X_train,y_train)
```

```python
y_pred = L_reg.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mse_Linear = mean_squared_error(y_test, y_pred)
print("Testing MSE:", mse_Linear)
```

```python
mae_Linear = mean_absolute_error(y_test, y_pred)
print("Testing MAE:", mae_Linear)
```

```python
#R_squred for linear regression
L_reg.score(X_train,y_train)
```

```python
L_reg.score(X_test,y_test)
```

```python

plt.scatter(X_train['Gender'], y_train, color='blue', label='Gender vs Weight')

plt.scatter(X_train['Height'], y_train, color='green', label='Height vs Weight')
plt.xlabel('Gender and Height')
plt.ylabel('Weight')
plt.title('Scatter Plot of Gender and Height vs Weight')
plt.legend()
plt.grid(True)
plt.show()
```
