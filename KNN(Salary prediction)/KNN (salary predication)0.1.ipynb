KNN

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
x=df[['age']]
y=df[['income']]
```

```python
knn=KNeighborsRegressor(n_neighbors=3)
knn.fit(x,y)
```

```python
knn_score=knn.score(x,y)
knn_score
```

```python
pred_salary_knn=knn.predict([[22]])
pred_salary_knn
```

Linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
LR=LinearRegression()
```

```python
LR.fit(x,y)
```

```python
pred_salary_lr=LR.predict([[22]])
pred_salary_lr
```

```python
print("predicate valu using KNN:",pred_salary_knn)
print("predicate valu using LR:",pred_salary_lr)
```

```python
LR.score(x,y)
```
